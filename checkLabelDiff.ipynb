```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_1100.csv")
```

```python
malware_family_list = ["adware", "flooder", "ransomware", "dropper", "spyware", "packed", "crypto_miner", "file_infector", "installer", "worm", "downloader"]
```

```python
i = 1
for index, row in df.iloc[:].iterrows():
    if(row.malware_type != row.max_val_mal):
        print(i, index, row.sha256, row.malware_type, row.max_val_mal)
        i = i + 1
```

```python
max_val_mal = df[malware_family_list].idxmax(axis=1)
df.insert(2, "max_val_mal", max_val_mal)
```
